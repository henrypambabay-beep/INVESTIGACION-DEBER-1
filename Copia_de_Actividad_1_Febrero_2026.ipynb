````python
# ------------------------------------------------------------------------------
# CELDA 2: C√ìDIGO (GENERACI√ìN Y CARGA DEL DATASET)
# ------------------------------------------------------------------------------
import pandas as pd
import numpy as np

# --- Generaci√≥n de Datos Sint√©ticos ---
# Para garantizar la reproducibilidad, fijamos una semilla aleatoria.
np.random.seed(42)

# N√∫mero de pacientes por grupo
n_pacientes = 30

# Creaci√≥n de los grupos
grupos = ['Placebo'] * n_pacientes + ['Estandarol'] * n_pacientes + ['TensioRedux'] * n_pacientes

# Generaci√≥n de IDs de paciente
paciente_ids = [f'P{i:03d}' for i in range(1, n_pacientes * 3 + 1)]

# Simulaci√≥n de la Presi√≥n Arterial Sist√≥lica (PAS) inicial
# Asumimos que todos los pacientes parten de una condici√≥n similar de hipertensi√≥n leve.
pas_inicial_placebo = np.random.normal(loc=155, scale=8, size=n_pacientes)
pas_inicial_estandarol = np.random.normal(loc=154, scale=8, size=n_pacientes)
pas_inicial_tensioredux = np.random.normal(loc=156, scale=8, size=n_pacientes)
pas_inicial = np.concatenate([pas_inicial_placebo, pas_inicial_estandarol, pas_inicial_tensioredux])

# Simulaci√≥n de la reducci√≥n de PAS (el efecto del tratamiento)
# Placebo: efecto m√≠nimo, centrado en 2 mmHg.
reduccion_placebo = np.random.normal(loc=2, scale=3, size=n_pacientes)
# Estandarol: efecto moderado, centrado en 10 mmHg.
reduccion_estandarol = np.random.normal(loc=10, scale=4, size=n_pacientes)
# TensioRedux: efecto superior, centrado en 15 mmHg.
reduccion_tensioredux = np.random.normal(loc=15, scale=5, size=n_pacientes)

# Calculamos la PAS final
pas_final_placebo = pas_inicial_placebo - reduccion_placebo
pas_final_estandarol = pas_inicial_estandarol - reduccion_estandarol
pas_final_tensioredux = pas_inicial_tensioredux - reduccion_tensioredux
pas_final = np.concatenate([pas_final_placebo, pas_final_estandarol, pas_final_tensioredux])

# Creaci√≥n del DataFrame
df = pd.DataFrame({
    'paciente_id': paciente_ids,
    'grupo_tratamiento': grupos,
    'PAS_inicial': pas_inicial.round(1),
    'PAS_final': pas_final.round(1)
})

# Mezclar el DataFrame para que los grupos no aparezcan en orden
df = df.sample(frac=1, random_state=42).reset_index(drop=True)

# --- Muestra de los datos cargados ---
print("Datos del ensayo cl√≠nico cargados correctamente.")
print("A continuaci√≥n se muestran las primeras 5 filas del dataset:")
display(df.head())
````

# Preparar los datos: Calcular la variable de inter√©s.

````python
# --- EJERCICIO 1: PREPARACI√ìN DE DATOS Y AN√ÅLISIS DESCRIPTIVO ---

# Aseguramos que pandas y numpy est√©n importados
import pandas as pd
import numpy as np

# --- Generaci√≥n de Datos Sint√©ticos (Copiado de la celda 2 para asegurar que df exista) ---
# Para garantizar la reproducibilidad, fijamos una semilla aleatoria.
np.random.seed(42)

# N√∫mero de pacientes por grupo
n_pacientes = 30

# Creaci√≥n de los grupos
grupos = ['Placebo'] * n_pacientes + ['Estandarol'] * n_pacientes + ['TensioRedux'] * n_pacientes

# Generaci√≥n de IDs de paciente
paciente_ids = [f'P{i:03d}' for i in range(1, n_pacientes * 3 + 1)]

# Simulaci√≥n de la Presi√≥n Arterial Sist√≥lica (PAS) inicial
# Asumimos que todos los pacientes parten de una condici√≥n similar de hipertensi√≥n leve.
pas_inicial_placebo = np.random.normal(loc=155, scale=8, size=n_pacientes)
pas_inicial_estandarol = np.random.normal(loc=154, scale=8, size=n_pacientes)
pas_inicial_tensioredux = np.random.normal(loc=156, scale=8, size=n_pacientes)
pas_inicial = np.concatenate([pas_inicial_placebo, pas_inicial_estandarol, pas_inicial_tensioredux])

# Simulaci√≥n de la reducci√≥n de PAS (el efecto del tratamiento)
# Placebo: efecto m√≠nimo, centrado en 2 mmHg.
reduccion_placebo = np.random.normal(loc=2, scale=3, size=n_pacientes)
# Estandarol: efecto moderado, centrado en 10 mmHg.
reduccion_estandarol = np.random.normal(loc=10, scale=4, size=n_pacientes)
# TensioRedux: efecto superior, centrado en 15 mmHg.
reduccion_tensioredux = np.random.normal(loc=15, scale=5, size=n_pacientes)

# Calculamos la PAS final
pas_final_placebo = pas_inicial_placebo - reduccion_placebo
pas_final_estandarol = pas_inicial_estandarol - reduccion_estandarol
pas_final_tensioredux = pas_inicial_tensioredux - reduccion_tensioredux
pas_final = np.concatenate([pas_final_placebo, pas_final_estandarol, pas_final_tensioredux])

# Creaci√≥n del DataFrame
df = pd.DataFrame({
    'paciente_id': paciente_ids,
    'grupo_tratamiento': grupos,
    'PAS_inicial': pas_inicial.round(1),
    'PAS_final': pas_final.round(1)
})

# Mezclar el DataFrame para que los grupos no aparezcan en orden
df = df.sample(frac=1, random_state=42).reset_index(drop=True)

print("DataFrame 'df' generado y listo para su uso.")

# 1. Preparar los datos: Calcular la variable de inter√©s.
# La variable de inter√©s es la reducci√≥n de la Presi√≥n Arterial Sist√≥lica (PAS).
df['reduccion_PAS'] = df['PAS_inicial'] - df['PAS_final']

print("DataFrame con la nueva columna 'reduccion_PAS':")
display(df.head())

# 2. An√°lisis Descriptivo: Resumir los datos de cada grupo.
print("\nAn√°lisis estad√≠stico descriptivo de 'reduccion_PAS' por grupo de tratamiento:")
display(df.groupby('grupo_tratamiento')['reduccion_PAS'].describe())

# Visualizaci√≥n general: Histograma de la reducci√≥n de PAS por grupo
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='reduccion_PAS', hue='grupo_tratamiento', kde=True, palette='viridis')
plt.title('Distribuci√≥n de la Reducci√≥n de PAS por Grupo de Tratamiento')
plt.xlabel('Reducci√≥n de PAS (mmHg)')
plt.ylabel('Frecuencia')
plt.show()
````

**Verificaci√≥n de Supuestos: Comprobar si los datos cumplen las condiciones para realizar un ANOVA.**

````python
# ------------------------------------------------------------------------------
# EJERCICIO 1: VERIFICACI√ìN DE SUPUESTOS (Normalidad con Shapiro-Wilk)
# ------------------------------------------------------------------------------
from scipy.stats import shapiro

print("\n--- Verificaci√≥n de Supuestos: Normalidad (Prueba de Shapiro-Wilk) ---")
print("Analizando la distribuci√≥n de 'reduccion_PAS' por grupo de tratamiento:")

# Obtener los nombres √∫nicos de los grupos de tratamiento
grupos_unicos = df['grupo_tratamiento'].unique()

# Iterar sobre cada grupo y realizar la prueba de Shapiro-Wilk
for grupo in grupos_unicos:
    # Filtrar el DataFrame para obtener los datos del grupo actual
    data_grupo = df[df['grupo_tratamiento'] == grupo]['reduccion_PAS']

    # Realizar la prueba de Shapiro-Wilk
    statistic, p_value = shapiro(data_grupo)

    # Imprimir los resultados para el grupo actual
    print(f"\nGrupo: {grupo}")
    print(f"  Estad√≠stico Shapiro-Wilk: {statistic:.4f}")
    print(f"  Valor p: {p_value:.4f}")

    # Interpretar el resultado (basado en un nivel de significancia alpha = 0.05)
    if p_value > 0.05:
        print("  Conclusi√≥n: Los datos parecen seguir una distribuci√≥n normal (p > 0.05).")
    else:
        print("  Conclusi√≥n: Los datos NO parecen seguir una distribuci√≥n normal (p <= 0.05).")

print("\nNota: Un valor p > 0.05 sugiere que no hay evidencia suficiente para rechazar la normalidad.")
````

### Resultados de la Prueba de Normalidad (Shapiro-Wilk)

Hemos realizado la prueba de Shapiro-Wilk para verificar el supuesto de normalidad en la distribuci√≥n de la variable `reduccion_PAS` para cada uno de los grupos de tratamiento. Este paso es crucial antes de proceder con un an√°lisis param√©trico como el ANOVA.

Los resultados obtenidos son los siguientes:

*   **Grupo Estandarol:** Estad√≠stico Shapiro-Wilk = 0.9565, Valor p = 0.2512.
*   **Grupo Placebo:** Estad√≠stico Shapiro-Wilk = 0.9473, Valor p = 0.1435.
*   **Grupo TensioRedux:** Estad√≠stico Shapiro-Wilk = 0.9560, Valor p = 0.2447.

**Interpretaci√≥n:**

Para los tres grupos (Estandarol, Placebo y TensioRedux), el valor p obtenido en la prueba de Shapiro-Wilk es mayor que el nivel de significancia convencional (Œ± = 0.05). Esto significa que **no tenemos evidencia estad√≠stica suficiente para rechazar la hip√≥tesis nula** de que los datos provienen de una distribuci√≥n normal.

**Conclusi√≥n:**

Bas√°ndonos en los resultados de la prueba de Shapiro-Wilk, podemos considerar que el supuesto de normalidad se cumple razonablemente para la variable `reduccion_PAS` en cada uno de los grupos de tratamiento. Esto nos permite continuar con el An√°lisis de Varianza (ANOVA) como prueba estad√≠stica principal para comparar las medias de reducci√≥n de PAS entre los grupos.

**An√°lisis Inferencial Y Post Hoc**

````python
# ------------------------------------------------------------------------------
# EJERCICIO 1: AN√ÅLISIS INFERENCIAL COMPLETO (ANOVA con statsmodels y Post-Hoc Tukey)
# Colocado despu√©s del an√°lisis de normalidad seg√∫n solicitud.
# ------------------------------------------------------------------------------
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Re-creating df for standalone execution of this cell, ensuring it has the necessary data.
# This block is duplicated from cell FOwCvcw1oSse and others.
np.random.seed(42)
n_pacientes = 30
grupos = ['Placebo'] * n_pacientes + ['Estandarol'] * n_pacientes + ['TensioRedux'] * n_pacientes
paciente_ids = [f'P{i:03d}' for i in range(1, n_pacientes * 3 + 1)]
pas_inicial_placebo = np.random.normal(loc=155, scale=8, size=n_pacientes)
pas_inicial_estandarol = np.random.normal(loc=154, scale=8, size=n_pacientes)
pas_inicial_tensioredux = np.random.normal(loc=156, scale=8, size=n_pacientes)
pas_inicial = np.concatenate([pas_inicial_placebo, pas_inicial_estandarol, pas_inicial_tensioredux])
reduccion_placebo = np.random.normal(loc=2, scale=3, size=n_pacientes)
reduccion_estandarol = np.random.normal(loc=10, scale=4, size=n_pacientes)
reduccion_tensioredux = np.random.normal(loc=15, scale=5, size=n_pacientes)
pas_final_placebo = pas_inicial_placebo - reduccion_placebo
pas_final_estandarol = pas_inicial_estandarol - reduccion_estandarol
pas_final_tensioredux = pas_inicial_tensioredux - reduccion_tensioredux
pas_final = np.concatenate([pas_final_placebo, pas_final_estandarol, pas_final_tensioredux])
df = pd.DataFrame({
    'paciente_id': paciente_ids,
    'grupo_tratamiento': grupos,
    'PAS_inicial': pas_inicial.round(1),
    'PAS_final': pas_final.round(1)
})
df = df.sample(frac=1, random_state=42).reset_index(drop=True)
df['reduccion_PAS'] = df['PAS_inicial'] - df['PAS_final']


print("\n--- An√°lisis Inferencial: ANOVA Unidireccional ---")
print("Descomposici√≥n de Varianzas y Estad√≠stico F")

# Crear un modelo lineal usando la f√≥rmula de Wilkinson
model = ols('reduccion_PAS ~ C(grupo_tratamiento)', data=df).fit()

# Realizar el ANOVA y obtener la tabla de resultados
anova_table = sm.stats.anova_lm(model, typ=2)

# Imprimir la tabla ANOVA
print(anova_table)

# ------------------------------------------------------------------------------
# AN√ÅLISIS POST-HOC (si el ANOVA es estad√≠sticamente significativo)
# ------------------------------------------------------------------------------
alpha = 0.05
if anova_table['PR(>F)'][0] < alpha: # Check p-value for the group effect
    print("\n--- An√°lisis Post-Hoc (Tukey HSD) ---")
    print("El ANOVA fue estad√≠sticamente significativo, procediendo con pruebas post-hoc.")

    # Realizar la prueba post-hoc de Tukey HSD
    tukey_results = pairwise_tukeyhsd(endog=df['reduccion_PAS'], groups=df['grupo_tratamiento'], alpha=alpha)

    # Imprimir los resultados de Tukey HSD
    print(tukey_results)

    # Visualizar los resultados de Tukey HSD
    tukey_results.plot_simultaneous()
    plt.title('Resultados de la Prueba Post-Hoc de Tukey HSD')
    plt.show()

else:
    print(f"\nEl ANOVA no fue estad√≠sticamente significativo (p >= {alpha}).")
    print("No es necesario realizar pruebas post-hoc.")
````

### Interpretaci√≥n Acad√©mica Completa: ANOVA y An√°lisis Post-Hoc (Tukey HSD)

Como resultado del an√°lisis de varianza unidireccional, hemos examinado la influencia del factor `grupo_tratamiento` sobre la variable dependiente `reduccion_PAS`. Este an√°lisis nos permite determinar si existen diferencias estad√≠sticamente significativas en la media de reducci√≥n de la presi√≥n arterial sist√≥lica entre los tres grupos de tratamiento: Placebo, Estandarol y TensioRedux.

**Resultados del ANOVA:**

La tabla ANOVA detalla la descomposici√≥n de la variabilidad total observada en la reducci√≥n de la presi√≥n arterial en componentes atribuibles a las diferencias entre grupos (`C(grupo_tratamiento)`) y a la variabilidad residual dentro de los grupos (`Residual`):

| Source               | Sum Sq        | df  | Mean Sq       | F           | P-value     |
| :------------------- | :------------ | :-- | :------------ | :---------- | :---------- |
| `C(grupo_tratamiento)` | 3137.96       | 2.0 | 1568.98       | 101.20      | 1.97e-23    |
| `Residual`           | 1348.89       | 87.0 | 15.50        |             |             |

El estad√≠stico F observado para el efecto del `grupo_tratamiento` fue **101.20**, con 2 y 87 grados de libertad, y un valor p asociado de **1.97e-23**.

Dado que el valor p (1.97e-23) es extremadamente peque√±o y considerablemente menor que nuestro nivel de significancia pre-especificado (Œ± = 0.05), **rechazamos enf√°ticamente la hip√≥tesis nula**. Esto proporciona evidencia estad√≠stica robusta para concluir que existe una diferencia estad√≠sticamente significativa en la reducci√≥n promedio de PAS entre al menos uno de los pares de grupos de tratamiento. La magnitud del estad√≠stico F subraya que la variabilidad explicada por las diferencias entre grupos es mucho mayor que la variabilidad residual dentro de los grupos.

**Resultados del An√°lisis Post-Hoc (Tukey HSD):**

Dado que el resultado del ANOVA fue estad√≠sticamente significativo (p < 0.05), procedimos a realizar pruebas post-hoc utilizando el m√©todo de Tukey HSD para determinar entre qu√© pares espec√≠ficos de grupos existen estas diferencias significativas, controlando la tasa de error tipo I para comparaciones m√∫ltiples.

**Reflexi√≥n : Utilidad Comparativa de los Tratamientos**



Desde una perspectiva rigurosamente acad√©mica, los resultados del an√°lisis post-hoc de Tukey HSD (previamente ejecutado y con resultados en la tabla adjunta en el cuaderno) son inequ√≠vocos respecto a la utilidad comparativa de los tratamientos. La reducci√≥n media de PAS fue significativamente mayor en el grupo TensioRedux en comparaci√≥n con Estandarol (p < 0.05) y, de manera m√°s marcada, en comparaci√≥n con Placebo (p < 0.05). Asimismo, Estandarol demostr√≥ una reducci√≥n significativamente superior a Placebo (p < 0.05).

Por lo tanto, bas√°ndonos estrictamente en la magnitud y significancia estad√≠stica de la reducci√≥n de PAS observada en este ensayo, **TensioRedux emerge como el tratamiento m√°s √∫til** de los tres evaluados, demostrando una eficacia superior tanto al placebo como al tratamiento est√°ndar.
````python
# ------------------------------------------------------------------------------
# CELDA 5: C√ìDIGO (CARGA DEL DATASET - ACTIVIDAD 2)
# ------------------------------------------------------------------------------
# --- URL del dataset desde el repositorio de UCI ---
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data'

# --- Nombres de las columnas (seg√∫n la documentaci√≥n del dataset) ---
column_names = [
    'age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg',
    'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target'
]

# --- Carga de datos usando pandas ---
# El dataset usa '?' para los valores faltantes. Los cargamos como NaN para poder manejarlos.
df_act2 = pd.read_csv(url, header=None, names=column_names, na_values='?')

# --- Preprocesamiento B√°sico ---
# 1. Eliminar filas con cualquier valor faltante (es una estrategia simple y efectiva para este dataset)
df_act2.dropna(inplace=True)

# 2. Convertir la variable objetivo a un formato binario
# Originalmente: 0 = sin enfermedad, 1-4 = diferentes grados de enfermedad.
# Tarea: convertir a 0 = sin enfermedad, 1 = con enfermedad.
df_act2['target'] = (df_act2['target'] > 0).astype(int)

# --- Muestra de los datos cargados ---
print("\n--- ACTIVIDAD 2 ---")
print("Datos de 'Heart Disease UCI' cargados y preprocesados correctamente.")
print("A continuaci√≥n se muestran las primeras 5 filas del dataset:")
display(df_act2.head())
print("\nInformaci√≥n del DataFrame resultante:")
df_act2.info()
````

# Divisi√≥n de Datos: Separar√°s los datos en conjuntos de entrenamiento y prueba

````python
# ------------------------------------------------------------------------------
# EJERCICIO 2: DIVISI√ìN DE DATOS (Entrenamiento y Prueba)
# ------------------------------------------------------------------------------
from sklearn.model_selection import train_test_split

# Separar las caracter√≠sticas (X) y la variable objetivo (y)
X = df_act2.drop('target', axis=1)
y = df_act2['target']

# Dividir los datos en conjuntos de entrenamiento y prueba
# Usamos un tama√±o de prueba del 30% (test_size=0.3) y una semilla aleatoria (random_state=42)
# para asegurar la reproducibilidad de la divisi√≥n.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y) # stratify ensures similar proportion of target in train and test

print("\n--- Divisi√≥n de Datos ---")
print(f"Tama√±o total del dataset: {df_act2.shape[0]} filas")
print(f"Tama√±o del conjunto de entrenamiento (X_train, y_train): {X_train.shape[0]} filas")
print(f"Tama√±o del conjunto de prueba (X_test, y_test): {X_test.shape[0]} filas")
print("\nProporci√≥n de la variable objetivo en el conjunto original:")
print(y.value_counts(normalize=True))
print("\nProporci√≥n de la variable objetivo en el conjunto de entrenamiento:")
print(y_train.value_counts(normalize=True))
print("\nProporci√≥n de la variable objetivo en el conjunto de prueba:")
print(y_test.value_counts(normalize=True))
````

**Modelado**

````python
# ------------------------------------------------------------------------------
# EJERCICIO 2: MODELADO (An√°lisis Lineal Discriminante - LDA)
# ------------------------------------------------------------------------------
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns

print("\n--- Entrenamiento del Modelo de An√°lisis Lineal Discriminante (LDA) ---")

# Inicializar el modelo LDA
lda_model = LinearDiscriminantAnalysis()

# Entrenar el modelo con los datos de entrenamiento
lda_model.fit(X_train, y_train)

print("Modelo LDA entrenado correctamente.")

# Ahora, tambi√©n entrenaremos el modelo de Regresi√≥n Log√≠stica en la misma celda
print("\n--- Entrenamiento del Modelo de Regresi√≥n Log√≠stica ---")

# Inicializar el modelo de Regresi√≥n Log√≠stica
logistic_model = LogisticRegression(max_iter=1000, random_state=42) # Increased max_iter for convergence

# Entrenar el modelo con los datos de entrenamiento
logistic_model.fit(X_train, y_train)

print("Modelo de Regresi√≥n Log√≠stica entrenado correctamente.")
````

***Evalucion de modelos***

````python
# ------------------------------------------------------------------------------
# EJERCICIO 2: EVALUACI√ìN DE MODELOS (LDA y Regresi√≥n Log√≠stica)
# ------------------------------------------------------------------------------

print("\n--- Evaluaci√≥n de Modelos ---")

# --- Predicciones en el conjunto de prueba ---
y_pred_lda = lda_model.predict(X_test)
y_pred_logistic = logistic_model.predict(X_test)

# --- M√©tricas de Evaluaci√≥n ---

def evaluate_model(y_true, y_pred, model_name):
    """Calcula e imprime m√©tricas de evaluaci√≥n para un modelo."""
    print(f"\nResultados para el modelo: {model_name}")
    print("-" * (len(model_name) + 25))

    # Matriz de Confusi√≥n
    cm = confusion_matrix(y_true, y_pred)
    print("Matriz de Confusi√≥n:")
    print(cm)

    # Visualizar Matriz de Confusi√≥n
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
                xticklabels=['Predicho Negativo (0)', 'Predicho Positivo (1)'],
                yticklabels=['Real Negativo (0)', 'Real Positivo (1)'])
    plt.title(f'Matriz de Confusi√≥n - {model_name}')
    plt.xlabel('Predicci√≥n')
    plt.ylabel('Valor Real')
    plt.show()

    # Precisi√≥n (Accuracy)
    accuracy = accuracy_score(y_true, y_pred)
    print(f"Precisi√≥n (Accuracy): {accuracy:.4f}")

    # Precisi√≥n (Precision) - Capacidad del modelo para no etiquetar como positivo un negativo real
    # Calculamos la precisi√≥n para la clase positiva (1)
    precision = precision_score(y_true, y_pred, pos_label=1)
    print(f"Precisi√≥n (Precision) para Clase Positiva (1): {precision:.4f}")

    # Sensibilidad (Recall) - Capacidad del modelo para encontrar todas las muestras positivas reales
    # Calculamos la sensibilidad para la clase positiva (1)
    recall = recall_score(y_true, y_pred, pos_label=1)
    print(f"Sensibilidad (Recall) para Clase Positiva (1): {recall:.4f}")


# Evaluar el modelo LDA
evaluate_model(y_test, y_pred_lda, "An√°lisis Lineal Discriminante (LDA)")

# Evaluar el modelo de Regresi√≥n Log√≠stica
evaluate_model(y_test, y_pred_logistic, "Regresi√≥n Log√≠stica")

# --- Comparaci√≥n Adicional: Curvas ROC y AUC ---
print("\n--- Comparaci√≥n de Curvas ROC ---")

# Obtener las probabilidades de predicci√≥n para la clase positiva (1)
y_prob_lda = lda_model.predict_proba(X_test)[:, 1]
y_prob_logistic = logistic_model.predict_proba(X_test)[:, 1]

# Calcular la curva ROC y el AUC para LDA
fpr_lda, tpr_lda, _ = roc_curve(y_test, y_prob_lda)
auc_lda = auc(fpr_lda, tpr_lda)

# Calcular la curva ROC y el AUC para Regresi√≥n Log√≠stica
fpr_logistic, tpr_logistic, _ = roc_curve(y_test, y_prob_logistic)
auc_logistic = auc(fpr_logistic, tpr_logistic)

# Plotear las curvas ROC
plt.figure(figsize=(8, 6))
plt.plot(fpr_lda, tpr_lda, label=f'LDA (AUC = {auc_lda:.4f})')
plt.plot(fpr_logistic, tpr_logistic, label=f'Regresi√≥n Log√≠stica (AUC = {auc_logistic:.4f})')
plt.plot([0, 1], [0, 1], 'k--', label='Aleatorio (AUC = 0.50)') # Curva aleatoria
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
plt.title('Curva ROC (Receiver Operating Characteristic)')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()

print(f"\n√Årea Bajo la Curva ROC (AUC) para LDA: {auc_lda:.4f}")
print(f"√Årea Bajo la Curva ROC (AUC) para Regresi√≥n Log√≠stica: {auc_logistic:.4f}")
````
